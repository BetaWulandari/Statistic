```
#membaca file dataset csv
#profil dataset : Sebuah penelitian komputer dan ram.
data=pd.read_csv("ramkomputer.csv",encoding='latin1')
#menampilkan 20 data teratas
data.head(20)
```

```
#melihat struktur dataset
data.info()
```

```
import numpy as np
# menentukan dataframe
hitung = data['ram']
# nilai modus
modus = hitung.mode()
print('Modusnya adalah : ',modus[0])
```

median


```
# mencari median
median = hitung.median()
median
```

rata-rata

```
#mencari rata-rata
rata = hitung.mean()
rata
```

```
# mencari ukuran sebaran data
maks = hitung.max()
minm = hitung.min()
range = maks - minm
print ('Nilai range adalah :',range)
```

simpangan baku

```
standardeviasi=hitung.std()
standardeviasi
```

visualisasi

```
#libary
import matplotlib.pyplot as plt
# membuat grafik
plt.figure(figsize=(15,5))
plt.hist(hitung,bins=35,color='grey')
plt.axvline(rata,color='red',label='Mean')
plt.axvline(median,color='yellow',label='Median')
plt.axvline(modus[0],color='green',label='Modus')
plt.xlabel('ram')
plt.ylabel('frekuensi')
plt.legend()
plt.show()
```
